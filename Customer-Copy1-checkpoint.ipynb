```python
from statsmodels.tsa.stattools import adfuller

def adf_test(dataset):
  dftest = adfuller(dataset, autolag = 'AIC')
  print("a. ADF : ",dftest[0])
  print("b. P-Value : ", dftest[1])
  print("c. Num Of Lags : ", dftest[2])
  print("d. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
  print("e. Critical Values :")
  for key, val in dftest[4].items():
      print("\t",key, ": ", val)
```
